<h2> IMPORTING LIBRARIES

<h2> EXPLORATORY DATA ANALYSIS

<h3> GENDER

```python
ax = sb.countplot(x='Gender',data=df)
for count in ax.containers:
    ax.bar_label(count)
```

<h5> INSIGHT: 
    Females have purchased more products than males at the time of Diwali season.

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
```

```python
sb.barplot(data=sales_gen,x='Gender',y='Amount')
```

<h5> INSIGHT: 
    Even the purchasing power of Females was higher than males at the time of Diwali season.

<h3> AGE

```python
ax = sb.countplot(data = df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sb.barplot(x='Age Group',y='Amount',data=sales_age)
```

<h5> INSIGHT: 
    Most number of buyers are females belonging to the age group of 26-35.

<h3> STATE

```python
# Total number of orders from Top 10 states
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
plt.figure(figsize=(20,8))
sb.barplot(x='State',y='Orders',data=sales_state)
```

```python
# Total sales/amount from top 10 states
sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
plt.figure(figsize=(20,8))
sb.barplot(x='State',y='Amount',data=sales_state)
```

<h5> INSIGHT: 
    Most of the orders and total sales/amount comes from Uttar Pradesh, Maharshtra & Karnatka.

<h3> MARITAL STATUS

```python
ax= sb.countplot(data=df,x='Marital Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_mstatus = df.groupby(['Marital Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(10,7))
sb.barplot(x='Marital Status',y='Amount',hue = 'Gender', data=sales_mstatus)
```

<h5> INSIGHT: 
    Most of the buyers are married women and their purchasing power is also higher.

<h3> OCCUPATION

```python
plt.figure(figsize=(20,8))
ax = sb.countplot(x='Occupation',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(20,7))
sb.barplot(x='Occupation',y='Amount', data=sales_occ)
```

<h5> INSIGHT: 
    Most of the buyers belong to IT, Healthcare and Aviation Sector.

<h3> PRODUCT CATEGORY


```python
plt.figure(figsize=(24,8))
ax = sb.countplot(x='Product Category',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Top 10 Product Categories by their total amount of sales
sales_prod = df.groupby(['Product Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
plt.figure(figsize=(20,8))
sb.barplot(x='Product Category',y='Amount', data=sales_prod)
```

<h5> INSIGHT: 
    Most of the sold products belong to Food, Clothing and Electronics Category.

```python
# Top 10 products by ID
sales_prodid = df.groupby(['Product ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
plt.figure(figsize=(20,8))
sb.barplot(x='Product ID',y='Amount', data=sales_prodid)
```

<h5> INSIGHT: 
    From the above graph these 3 Products have been the maximum number of sales - P00265242, P00110942 and P00184942.

<h3>OVERALL ANALYSIS

<h5>  GENERAL INSIGHT:
Married Women belonging to the age group of 26-35yrs from UP, Maharashtra and Karnataka working in IT, Healthcare and Aviation Sector are more likely to buy products from Food, Clothing and Electronics Category.
